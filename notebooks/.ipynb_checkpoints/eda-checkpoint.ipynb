```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("students.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=["Student_ID"], inplace=True)
```

## Handling Missing Values

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
```

```python
df["Gender"] = df["Gender"].fillna(df["Gender"].mode()[0])
```

```python
df = df[(df["Age"] >= 18) & (df["Age"] <= 23)]
```

```python
df["Gender"] = df["Gender"].str.strip().str.lower()
```

```python
df.drop_duplicates(inplace=True)
```

## EDA

```python
placement_counts = df["Placement_Status"].value_counts()
print("\nPlacement Distribution:")
print(placement_counts)
placement_counts.plot(kind="bar")
plt.title("Placement Status Distribution")
plt.xlabel("Placement Status")
plt.ylabel("Count")
plt.show()
```

```python
cgpa_avg = df.groupby("Placement_Status")["CGPA"].mean()
print("\nAverage CGPA:")
print(cgpa_avg)
cgpa_avg.plot(kind="bar")
plt.title("Average CGPA by Placement")
plt.ylabel("Average CGPA")
plt.show()
```

```python
internship_avg = df.groupby("Placement_Status")["Internships"].mean()
print("\nAverage Internships:")
print(internship_avg)
internship_avg.plot(kind="bar")
plt.title("Internships vs Placement")
plt.ylabel("Average Internships")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=["int64", "float64"])
correlation = numeric_df.corr()
print("\nCorrelation Matrix:")
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
df["CGPA"].hist(bins=20)
plt.title("CGPA Distribution")
plt.xlabel("CGPA")
plt.ylabel("Frequency")
plt.show()
```

## Feature Engineering

```python
df["Gender"] = df["Gender"].map({ "male": 0, "female": 1})
```

```python
df["Placement_Status"] = df["Placement_Status"].map({ "Not Placed": 0,"Placed": 1})
```

```python
df = pd.get_dummies(df,columns=["Degree", "Branch"],drop_first=True)
```
